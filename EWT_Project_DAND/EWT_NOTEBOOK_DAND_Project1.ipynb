# Exploring Weather Trends Project¶
### This Jupiter notebook is performing following tasks:reading the data from CSV File and visualize it by using 'Moving Average'

#### Importing the liberaries required for this project :

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Importing the data!


```python
ds=pd.read_csv("AvgTempYearlyData.csv") #for importing datasets containing yearly average Temperature data for global and local city (Delhi) 
```

#### Defining Function  'simplifiedRollingMean' Caculation for the Global data and Local data.
###### Moving averages are used to smooth out data and to make it easier to observe long term trends and not get lost in daily fluctuations. 


```python
def simplifiedRollingMean(windowRolling, ds_i):
    ds_o = ds_i.rolling(window = windowRolling, center=False, on = "year").mean().dropna()
    return ds_o
```

##### Doing calculation using 'ds_movingAvg' in order to create multiple plots by just changing rollingWindow Value.

```python
rollingWindow = 100
ds_movingAvg = simplifiedRollingMean(rollingWindow, ds)
```

#### Draw Graph in MatplotLIB  :
##### For comparing the Global Temperature data with Local Temperature data¶

```python
plt.plot(ds_movingAvg['year'],ds_movingAvg['global_avg_temp'],label='Global')
plt.plot(ds_movingAvg['year'],ds_movingAvg['city_avg_temp'],label='Delhi')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (°C)") 
plt.title("Temperature in Delhi versus Global values ({} year moving avg)".format(rollingWindow))
plt.show()
```

#### Caclulating Correlation Cofficient Between Year and Temperature(°C) for Global Temperature Trends :

```python
np.corrcoef(ds.global_avg_temp,ds.year)
```

#### Draw a scatter plot between Year and Temperature(°C) for Global Temperature Trends

```python
plt.scatter(ds.global_avg_temp,ds.year)
plt.ylabel("Years")
plt.xlabel("Temperature (°C)") 
plt.title("Correlation Cofficient Between Year and Temperature(°C) for Global Temperature Trends")
plt.show()
```
